```python
#importing necessary python libraries and packages
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
#loading House Price India dataset which is in csv format
df = pd.read_csv("House Price India.csv")
df.head() #returns first 5 rows of the dataset
```

```python
df.shape  #gives number of rows, number of columns
```

```python
df.info() #returns number of non-null values and data type of each column
```

```python
x = df[["living area"]].values  #independent variable
y = df[["Price"]].values        #dependent variable
```

```python
model = LinearRegression()  #calling linear regression module
model.fit(x, y)   #assigning axis in the model
```

```python
y_pred = model.predict(x) #predicts y value
mse_original = mean_squared_error(y, y_pred)  #gives mean square error between original y and y predict
rmse_original = np.sqrt(mse_original) #gives root mean square error between original y and y predict
```

```python
plt.figure(figsize=(5, 5))
plt.scatter(x, y, color="skyblue", alpha=0.8, label="Actual Prices")
plt.plot(x, y_pred, color="red", linewidth=2, label="Regression Line")
plt.title("Linear Regression: Living Area vs Price")
plt.xlabel("Living Area (sq ft)")
plt.ylabel("Price (INR)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
print("Mean Square Error on Original Data: {:,.2f}".format(mse_original))
```

```python
print("Root Mean Square Error on Original Data: {:,.2f}".format(rmse_original))
```

```python
np.random.seed(42)
noise = np.random.normal(0, 100000, size=y.shape)  # Std dev = ₹1 Lakh
y_noisy = y + noise
```

```python
model_noisy = LinearRegression()
model_noisy.fit(x, y_noisy)
```

```python
y_pred_noisy = model_noisy.predict(x)
mse_noisy = mean_squared_error(y_noisy, y_pred_noisy)
rmse_noisy = np.sqrt(mse_noisy)
```

```python
plt.figure(figsize=(5, 5))
plt.scatter(x, y_noisy, color="lightgreen", alpha=0.8, label="Noisy Prices")
plt.plot(x, y_pred_noisy, color="yellow", linewidth=2, label="Regression Line (Noisy Data)")
plt.title("Linear Regression with Noisy Data")
plt.xlabel("Living Area (sq ft)")
plt.ylabel("Price (INR)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
print("Mean Square Error on Noisy Data: {:,.2f}".format(mse_noisy))
```

```python
print("Root Mean Square Error on Original Data: {:,.2f}".format(rmse_noisy))
```

```python
print("Mean Square Error on Original Data: {:,.2f}".format(mse_original))
print("Mean Square Error on Noisy Data: {:,.2f}".format(mse_noisy))
print("Root Mean Square Error on Original Data: {:,.2f}".format(rmse_original))
print("Root Mean Square Error on Noisy Data: {:,.2f}".format(rmse_noisy))
```

Mean square error increased after noise is added
