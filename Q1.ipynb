```python
#importing necessary python libraries
import numpy as np
import pandas as pd
```

```python
#loading the dataset which is in csv format using pandas
df = pd.read_csv('student_performance.csv')
df.head()   #default head gives first 5 rows of the dataset
```

```python
df.shape  #returns number of rows , number of columns
```

```python
df.info() #returns number of non-null values and data type of each column
```

```python
#calculating total number of students in the given dataset
Total_number_of_students = len(df['student_id'])  #number of student ids gives total number of students in the given dataset
Total_number_of_students
```

```python
#calculating number of students who studied more than 2 hours
Students_studied_more_than_2hours = df[df['study_hours_per_day'] > 2]   #returns the dataframe of students who studied more than 2 hours
Count_Students_studied_more_than_2hours = len(Students_studied_more_than_2hours)  #length of that dataframe gives the required count
Count_Students_studied_more_than_2hours
```

```python
#calculating number of students who passed the test
Students_passed = df[df['exam_score'] >= 35]   #returns the dataframe of students who got more than or equal to 35 marks
Count_Students_passed = len(Students_passed)   #length of that dataframe gives the required count
Count_Students_passed
```

```python
#claculating number of students who studied > 2 hrs and passed the test
Students_passed_and_studied_more_than_2hours = Students_studied_more_than_2hours [Students_studied_more_than_2hours ['exam_score'] >= 33] #returns the required dataframe
Count_Students_passed_and_studied_more_than_2hours = len(Students_passed_and_studied_more_than_2hours)  #length of that dataframe gives the required count
Count_Students_passed_and_studied_more_than_2hours
```

```python
#calculating percentage of students who passed among those who studied > 2 hours
Required_percentge = (Count_Students_passed_and_studied_more_than_2hours/Total_number_of_students)*100
Required_percentge
```

```python
#Using Bayes Theorem to compute: Probability that a student studied > 2h given they passed
Probability = (Count_Students_passed_and_studied_more_than_2hours/Count_Students_passed)
Probability
```
